We can also define how many elements we want inside our figure. For that we need to specify nrows and ncolumns

What are Axes in this context?

 When you create subplots (e.g., nrows=2, ncols=2), Matplotlib generates a 2D array of Axes objects (each Axes represents a single subplot).

 You can think of axes as a collection (grid) of all the subplots inside the figure.

```python
ax1.plot(np.random.randn(50), c = "red", linestyle = "--")
ax2.plot(np.random.randn(50), c ="green", linestyle = ":")
ax3.plot(np.random.randn(50), c ="blue", marker="o" , linewidth=3.0)
ax4.plot(np.random.randn(50), c ="yellow")
fig

# Here the np.random.randn(50) will generate a 1D numpy array with 50 elements
# We can also pass the number of rows and columns as arguments in it for creating 2 D array. Similarly we pass three arguments it creates a 3D array
```

# The subplot2grid command

There is another way to make subplots using a grid like format

```python
plt.figure(figsize=(14,6))

#syntax = plt.subplot2grid(shape, location, rowspan=1, colspan=1)

# rowspan = how many rows tall your subplot will be.
# colspan = how many columns wide your subplot will be.

ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)
# This simply starts at top left cell and covers the 3 columns of the top row
ax2 = plt.subplot2grid((3,3), (1,0),colspan=2)
# This simply starts at the row 1 col 0 and takes 2 cols
ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2)
# This simply starts at row 1 col 2 and takes 2 rows .
ax4 = plt.subplot2grid((3,3), (2,0))
# This takes the row 2 0 column
ax5 = plt.subplot2grid((3,3), (2,1))
# This takes the row 2 column 1
```

The use cases of the above subplot2grid are as follows: 

 Financial dashboards showing price trends, volume, and indicators in one figure.

 Scientific reports where multiple types of graphs are shown with different sizes.

 Interactive apps or presentations needing neat, professional layouts.

# Scatter Plot 

# 1. What is a Scatter Plot?

A scatter plot is a graph where each data point is represented as a dot on a 2D plane, using:
 X-axis: Represents one variable.

 Y-axis: Represents another variable.

For example, if you have height vs weight data, each dot in the scatter plot will represent a persons height and weight.

# 2. Why are Scatter Plots Used?

Scatter plots are essential because they:

1.Show relationships (correlation) between two variables:

	 Are they positively related? (e.g., more study hours  higher marks)

	 Are they negatively related? (e.g., more distance from the city  lower rent prices)

	 No clear relationship?

2.Identify patterns/trends:

	 You can see if the data forms a linear, curved, or random pattern.

3.Find outliers:

	 Points that are far away from the rest of the data.

4.Compare multiple variables:

	 By using color or size of markers, you can show a 3rd variable.
	

```python
# Basic syntax : plt.scatter(x, y, color='blue', marker='o', s=50, alpha=0.7)
N = 50
x =np.random.randn(N)
y =np.random.randn(N)
colors = np.random.randn(N)
area = np.pi * (20 * np.random.randn(N)) ** 2
# this creates area for each of the elements in the array formed by using random numbers

# The above data are just for the demonstration purpose and the visualization will make more sense once we work with the real data
```

```python
plt.figure(figsize=(14,6))
plt.scatter(x,y, s=area,c=colors, alpha=0.5, cmap="Spectral")
plt.colorbar()
plt.show()

# This is creating a scatter plot with:
# 		x, y  Coordinates of the 50 points.
# 		s=area  Size of each point (bubble), which you calculated earlier.
# 		c=colors  Color of each point, based on a third variable (array of 50 values).
# 		alpha=0.5  Controls the transparency (or opacity) of the points.
# 		cmap="Spectral"  The colormap that maps your numeric colors to actual visible colors.
```

# The scatter plot helps you analyze four dimensions of variable in one figure.

```python
fig = plt.figure(figsize=(14,6))

ax1 = fig.add_subplot(1,2,1)
ax1.set_title("Number One")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
scatter1 = ax1.scatter(x,y, s=area,c=colors, alpha=0.5, cmap="Pastel1")

# cmap : Take the numbers in some_values, and convert them into colors using the color map named 'viridis'.

ax2 = fig.add_subplot(1,2,2)
ax2.set_title("Number Two")
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
scatter2 = ax2.scatter(x, y, s=area, c = colors, alpha=0.5, cmap="Pastel2")

fig.colorbar(scatter2, ax=[ax1, ax2], orientation='vertical')
# This is simply to provide the color index like the legend 
plt.show()
```

## Histogram

# What is a Histogram?

A histogram is a type of plot that shows the distribution of a single numeric variable by grouping data into bins (ranges) and showing how many values fall into each bin.

# Histogram = Binned Frequency Plot

 X-axis  Intervals or bins (e.g., age groups, score ranges)

 Y-axis  Frequency (how many data points fall in each bin)

 The higher the bar, the more values fall in that range.


# When is it used?

 Exploratory Data Analysis (EDA)  to understand the shape of your data

 To check for normal distribution before statistical tests

 In machine learning to understand feature scales

 For visualizing survey results, sales numbers, grades, etc.

```python
values = np.random.randn(1000)
```

```python
plt.subplots(figsize=(14,6))

plt.hist(values, bins = 100, alpha =0.7, histtype= "bar", color="steelblue", edgecolor="green")
plt.xlim(xmin=-5, xmax=5)
#The above xlim function only show the x-axis from -5 to +5, and ignore anything outside this range.
plt.show()
```

## Kernel Density Distribution

Kernel Density Estimation (KDE) is a non-parametric way to estimate the probability distribution of a continuous variable.
In simple terms, it helps you create a smooth curve that represents how your data is distributed  like a more elegant and flexible version of a histogram.

## What does KDE facilitate?

KDE allows you to:
	
     Understand how data values are spread
	
     Identify outliers or rare values
	
     Compare distributions between two or more groups
	
     Perform probability analysis without assuming a fixed model (like normal distribution)

A histogram might tell you that most students scored between 70 and 80, but a KDE curve shows the shape of the score distribution more clearly  like how smoothly the scores increase or decrease, and if theres a second small group of students who did really well around 90.

```python
from scipy.stats import gaussian_kde
density = gaussian_kde(values)
```

# Why use gaussian_kde and assign it to density?
This line creates a tool (function) that you can later use to:
	
    1. Estimate the probability density of any value
	
    2. Visualize the data distribution smoothly
	
    3. Compare multiple datasets
	
    4. Detect patterns that histograms may miss

Why is this important in Data Science?

Because real-world data is messy:
	
     Its rarely clean and perfectly shaped.
	
     Histograms are useful, but theyre sensitive to bin size and look jagged.
	
     KDE smooths that mess into a clean, readable curve.

# density(x) tells you how common or likely a value is in your dataset  like a probability without needing actual probability theory.

```python
plt.subplots(figsize = (14,6))
values2 = np.linspace(min(values) - 10, max(values) + 10, 100)

# This line creates 100 evenly spaced numbers between:
# 		min(values) - 10 (a bit below your smallest value)
# 		max(values) + 10 (a bit above your largest value)

#Purpose 
# To create a smooth range of x-values over which you can evaluate a function  like density(x) in a KDE plot.

plt.plot(values2, density(values2), color ="#87CEEB")
plt.fill_between(values2, 0, density(values2), alpha =0.5, color = "#87CEEB")
plt.xlim(xmin=-5, xmax =5)
plt.show()


# values2 = np.linspace(...)  # x-range
# density = gaussian_kde(...) # estimator
# plt.plot(...)               # curve
# plt.fill_between(...)       # shaded region
```

# Combined Plots

```python
plt.subplots(figsize = (14,6))

plt.hist(values, bins=100, color="steelblue", alpha=0.5, density=1, histtype="bar", edgecolor = "black")


plt.plot(values2, density(values2), color="#FF7F00", linewidth = 3.0)
plt.xlim(xmin=-5, xmax=5)
plt.show()

# The main purpose of the above code is to compare a histogram (bar-based view of distribution) with a KDE curve (smooth view of distribution) on the same plot.
```

## Bar Plots 

A Bar Plot (or Bar Chart) is a common visualization used to represent categorical data with rectangular bars. Each bars height (or length) corresponds to the value of the category it represents.

#  Uses of Bar Plots:
	 Show comparison between categories (e.g., sales by region)
	
     Display frequency counts or aggregated values
	
     Visualize grouped or stacked data
	
     Communicate ranking or distribution clearly

```python
Y = np.random.randn(1,5)[0]
# np.random.randn(1, 5) creates a 1 row  5 column 2D array of random numbers from a normal distribution (mean = 0, std = 1).
# Example: [[0.2, -0.5, 1.3, 0.0, -1.1]]
# [0] extracts the first row to make it a simple 1D array:
Y2 = np.random.randn(1,5)[0]
```

```python
plt.figure(figsize=(14, 6))
barWidth = 0.5

plt.bar(np.arange(len(Y)), Y, color = "green", label = "Y label")
plt.legend()
plt.show()
```

# Stacking two barplots 

```python
plt.figure(figsize=(14,6))

# Define the width of the bars
barWidth = 0.5

# Plot the first set of bars using Y values (e.g., red bars)
# np.arange(len(Y)) generates positions for each bar on the x-axis
# The x is being created using the numpy array creating fucntion till the length of the Y.
plt.bar(np.arange(len(Y)), Y, width=barWidth, color = "red", label ="Label Y")


# Plot the second set of bars using Y2 values (e.g., black bars)
# Using the same x positions, so this will **overlap** with the red bars
plt.bar(np.arange(len(Y2)),Y2, width = barWidth, color = "black", label = "Label Y2")

plt.legend()
plt.show()
```

## Box Plots

A Box Plot is a statistical chart used to visualize the distribution of a dataset by showing its summary statistics  including the median, quartiles, range, and outliers  in a single compact graphic.

## Uses of Box Plot:
	 Identify central tendency (median) of data
	
     Observe spread/variability (via interquartile range)
	
     Detect outliers (points outside whiskers)
	
     Compare distributions across multiple groups
	
     Quickly summarize large datasets visually

```python
values = np.concatenate([np.random.randn(10), np.array([10, 15, -10, -15])])
values
```

```python
plt.figure(figsize=(14,6))
plt.hist(values)
```

```python
plt.figure(figsize = (14,6))
plt.boxplot(values)

#Syntax : plt.boxplot(data, notch=False, vert=True, patch_artist=False, meanline=False, showmeans=False, showcaps=True, showbox=True, whiskerprops=None, boxprops=None, capprops=None, flierprops=None, medianprops=None, meanprops=None, widths=0.5)
```

In a box plot, the central box represents the middle 50% of the data (from the first quartile Q1 to the third quartile Q3).
Inside the box, the orange (or bold) line marks the median (Q2), which is the middle value of the dataset.


Outliers are not removed automatically  they are just flagged or shown separately.
Whether to remove them or not depends on your data cleaning strategy.
