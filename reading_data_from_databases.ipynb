# Reading data from databases  

In this lesson we will learn to read SQL queries and relational database table into DataFrame objects using pandas. 

```python
import pandas as pd 
```

Read data from SQL Database

Reading data from SQL relational database is fairly simple and pandas support a variety of methods to deal with it.
We will start with an example using SQLlite , as its a bulletin python package. Ww wont need anything extra installed.

In order to work with a SQLLite database from Python, we first have to connect it. We can do it by using the connect function which returns a connection object.

```python
import sqlite3

#This line imports Python’s built-in library for working with SQLite databases.
```

What is SQLite?

	• A lightweight, file-based SQL database.
	
    • No server needed — your whole database lives in a single .sqlite or .db file.
	
    • Perfect for small projects, data analysis, or embedded applications.

```python
conn = sqlite3.connect("dataset/chinook.db")

# This line connects your Python code to a SQLite database file named chinook.db.

# If the file already exists, it opens it.
# If it doesn’t exist, SQLite will create a new blank .db file named chinook.db.
```

Once we have the connection object we can create a Cursor Object. Cursor allows us to execute SQL Queries against a database.

```python
cur = conn.cursor() 
```

The sole purpose of creating a cursor in Python when working with SQL databases is to:

Execute SQL commands (like SELECT, INSERT, UPDATE, DELETE, etc.) and retrieve results if needed.

```python
cur.execute("SELECT name FROM sqlite_master WHERE type='table';")

tables = cur.fetchall()

for table in tables:
    print(table[0])

```

```python
cur.execute('SELECT * FROM Employee LIMIT 5;')
#It’s requesting 5 rows from the Employee table. This only sends a query
```

Similarly we have not assigned the results to one of the varaible because we will still need to run another command to fetch the result. We can use the **fetchall** query to fetch all the results of a query. 

```python
results = cur.fetchall()
```

```python
results
```

Since the above results are the list of tuples. Each tuple corresponds to the row in the database that we accessed. Dealing with data this way is painful. So pandas is there to help us with that. 

```python
df = pd.DataFrame(results)
```

```python
df.head()
```

Before we moveon its a good practise to close the connection objects and the cursor objects that are open. This prevents the SQLite database from being locked. When a SQLite database is locked , we may be unable to make update in the database, and may get errors. We can close the cursor and the connection like this.

```python
cur.close()
#Ends SQL execution channel
conn.close()
#Ends entire connection to the database file 
```

# Using pandas **read_sql** method

We can use the pandas reas_sql function directly to read the results of a SQL Query directly into the pandas DataFrame. The code below will execute the same query we did , but it will return us a dataframe. It has several advantages over the queries we did above in the code.

1. It doesnot require us to create a cursor object or call the fetchall at the end.

2. It automatically reads in the name of the header from the table.

3. It creates a DataFrame so we can quickly explore the data.

```python
conn = sqlite3.connect("dataset/chinook.db")
```

```python
df = pd.read_sql('SELECT * FROM Employee;', conn)
```

```python
df.head()
```

```python
df = pd.read_sql("SELECT * FROM Employee;", conn,
                 index_col = "EmployeeId",
                 parse_dates= ["BirthDate", "HireDate"])
```

```python
df.head()
```

```python
df.info()
```

```python
df["ReportsTo"].isna().sum()
```

```python
df["ReportsTo"].mean()
```

```python
df["ReportsTo"] > 2.85
```

```python
df["City"] = df["City"].astype("category")
#Fixed reapeating values being chenged to the data type category
```

```python
df.info()
```

# read_sql() acts like a smart shortcut:
	
    • If you pass it a SELECT query, it uses read_sql_query().
	
    • If you pass it a table name, it uses read_sql_table().
